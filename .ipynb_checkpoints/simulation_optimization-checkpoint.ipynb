```python
from elo import Elo
from glicko import Glicko
from glicko2 import Glicko2
import numpy as np
from func import prob_log
import matplotlib.pyplot as plt
import time
from scipy.stats import rankdata
```

```python
# Helper functions
def min_suffix_avg(arr):
    arr = np.asarray(arr)
    suffix_sums = np.cumsum(arr[::-1])[::-1]       # Suffix sums
    counts = np.arange(1, len(arr) + 1)[::-1]
    suffix_avgs = suffix_sums / counts
    
    min_index = np.argmin(suffix_avgs[0:len(suffix_avgs)])         # index of min average
    min_avg = suffix_avgs[min_index]
    min_avg_low = np.min(arr[min_index:])
    min_avg_high = np.max(arr[min_index:])
    print(suffix_avgs[0:len(suffix_avgs)])
    return min_index, min_avg, min_avg_low, min_avg_high

def find_convergence_start(arr, window=100, slope_thresh=0.001):
    arr = np.asarray(arr)
    n = len(arr)
    
    for i in range(n - window):
        # Fit a linear trend over the window
        y = arr[i:i+window]
        x = np.arange(window)
        coeffs = np.polyfit(x, y, 1)
        slope = coeffs[0]
        
        if abs(slope) < slope_thresh:
            return i, coeffs  # First point where it starts flattening
    
    return -1, None  # If no flat region is found
```

```python
# Initializing Elo Parameters

num_players = 1000
init_rating = np.full(num_players, 1500, dtype=np.float32)
init_true_rating = np.random.normal(1500, 350, num_players).astype(np.float32)
```

```python
# Initalizing Elo Simulations

elo = [] # elo[i] refers to the simulation with K constant = i+10
k_range = (10, 80)
k_len = k_range[1] - k_range[0]

# Creating the Elo simulations with K constants from 10 to 80 
for k_const in range(*k_range):
    elo.append(Elo(num_players, init_rating, init_true_rating, k_const))
```

```python
# Simulate Matches and Data Collection
start_time = time.time()

num_games = 1000

# Average absolute difference between true and visible ratings over time
avg_abs_diff_elo = [[np.mean(np.abs(elo[i].true_ratings - elo[i].ratings))] for i in range(k_len)]
avg_sq_diff_elo = [[np.mean(np.square(elo[i].true_ratings - elo[i].ratings))] for i in range(k_len)]

avg_ranking_dev = [[0] for _ in range(k_len)]
adj_period_ratings = [[np.copy(elo[i].ratings)] for i in range(k_len)]
# print(adj_period_ratings)

for i in range(k_len):
    # Ranking players by true rating for rank deviation data
    true_ranks = rankdata(elo[i].true_ratings, method='ordinal')                                         
    for j in range(num_games):
        # elo[i].match_all_perfect()
        elo[i].match_all_random()
        # elo[i].match_all_realistic()

        # Tracking avg absolute rating deviation from true ratings
        avg_abs_diff_elo[i].append(np.mean(np.abs(elo[i].true_ratings - elo[i].ratings)))
        avg_sq_diff_elo[i].append(np.mean(np.square(elo[i].true_ratings - elo[i].ratings)))
        
        if (j+1) % 50 == 0:
            # Ranking players by rating
            visible_ranks = rankdata(elo[i].ratings, method='ordinal')

            # Compute average ranking deviation
            ranking_deviation = np.abs(visible_ranks - true_ranks)
            avg_ranking_dev[i].append(np.mean(ranking_deviation))

        if j < 30:
            adj_period_ratings[i].append(np.copy(elo[i].ratings))
                
    
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
# Post Simulation Data Collection

avg_abs_diff_elo = np.array(avg_abs_diff_elo)
adj_period_ratings = np.array(adj_period_ratings)

# Average number of games for player to be within 25, 50, 75, 100 rating from their true rating
elo_avg_conv_25, elo_avg_conv_50 = np.argmax(avg_abs_diff_elo <= 25, axis=1), np.argmax(avg_abs_diff_elo <= 50, axis=1)
elo_avg_conv_75, elo_avg_conv_100 = np.argmax(avg_abs_diff_elo <= 75, axis=1), np.argmax(avg_abs_diff_elo <= 100, axis=1)


# elo_conv_stats[i] refers to K=i: 
# number of games for all players to converge, stationary avg deviation, lowest avg deviation, highest avg deviation
elo_conv_stats = []
elo_volatility_mse = []
for i in range(k_len):
    # Aggregate rating deviation convergence statistics
    # Slope and intercept will be used for volatility analysis
    conv_start, conv_coeffs = find_convergence_start(avg_abs_diff_elo[i], 100, 1e-3)
    elo_conv_stats.append(conv_start)
    
    # Volatility Analysis based on mean squared error
    if conv_start != -1:
        adj_x = np.arange(0, conv_start)
        adj_y = avg_abs_diff_elo[i, :conv_start]
        conv_x = np.arange(conv_start, num_games+1)
        conv_y = avg_abs_diff_elo[i, conv_start:]
        
        adj_coeffs = np.polyfit(adj_x, adj_y, 2)
        
        adj_pred = np.polyval(adj_coeffs, adj_x)
        conv_pred = np.polyval(conv_coeffs, conv_x)
    
        adj_residuals = adj_y - adj_pred
        conv_residuals = conv_y - conv_pred
    
        adj_ssr = np.sum(np.square(adj_residuals))
        conv_ssr = np.sum(np.square(conv_residuals))
    
        adj_mse = adj_ssr / (conv_start - 3) # -3 to account for degrees of freedom (quadratic regression)
        conv_mse = conv_ssr / (num_games - conv_start - 2) # -2 to account for linear regression

        elo_volatility_mse.append((adj_mse, conv_mse))
    else:
        elo_volatility_mse.append(-1)
        
elo_adj_vol = []
player_data = adj_period_ratings.transpose(0, 2, 1)
adj_x = np.arange(0, 31) # To include the 30th game played
for i in range(k_len):
    sum_mse = 0
    for j in range(num_players):
        adj_y = player_data[i, j, :]
        adj_coeffs = np.polyfit(adj_x, adj_y, 1)
        adj_pred = np.polyval(adj_coeffs, adj_x)
        adj_residuals = adj_y - adj_pred
        adj_ssr = np.sum(np.square(adj_residuals))
        adj_mse = adj_ssr / (31 - 2)
        sum_mse += adj_mse
    avg_mse = sum_mse / num_players
    elo_adj_vol.append(avg_mse)
    
```
